```python
import requests
from bs4 import BeautifulSoup
```

```python
data=requests.get(url='https://fbref.com/en/comps/9/Premier-League-Stats')
```

```python
soup=BeautifulSoup(data.text)
```

```python
sTable=soup.select('table.stats_table')
```

```python
sTable
```

```python
s=sTable[0]
```

```python
s=s.find_all('a')
```

```python
s=[l.get('href') for l in s]
```

```python
s=[l for l in s if '/squads/'in l]
```

```python
s
```

```python
links = [f"https://fbref.com{l}" for l in s]
```

```python
#data=requests.get(links[13])
#df=pd.read_html(data.text,match='Goal and Shot Creation')[0]
#df.head()
```

```python
import pandas as pd
import time
df_list=[]
for l in links:
    data=requests.get(l)
    df=pd.read_html(data.text,match='Goal and Shot Creation')[0]
    df_list.append(df)
    time.sleep(20)
df_final=pd.concat(df_list)
```

```python
df_final.columns=df_final.columns.droplevel()
```

```python
df_final.head()
```

```python
df_final.to_csv(r'Shot_Creation.csv',index=False)
```
